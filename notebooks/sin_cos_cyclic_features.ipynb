```python
import os
import pandas as pd
import numpy as np
```

```python
data_folder = "data"
filename = "train.csv"
df = pd.read_csv(os.path.join("..", data_folder, filename), parse_dates=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
df_subset = df[["date", "orders", "warehouse"]]
df_subset.head()
```

```python
df_subset.info()
```

```python
df_subset['dayofyear'] = df_subset['date'].dt.dayofyear
df_subset['month'] = df_subset['date'].dt.month
df_subset['dayofweek'] = df_subset['date'].dt.dayofweek
df_subset.head()
```

```python
# Create sin and cos features for day of year
df_subset['dayofyear_sin'] = np.sin(2 * np.pi * df_subset['date'].dt.dayofyear/365.25)
df_subset['dayofyear_cos'] = np.cos(2 * np.pi * df_subset['date'].dt.dayofyear/365.25)

# Create sin and cos features for month
df_subset['month_sin'] = np.sin(2 * np.pi * df_subset['date'].dt.month/12)
df_subset['month_cos'] = np.cos(2 * np.pi * df_subset['date'].dt.month/12)

# Create sin and cos features for day of week
df_subset['dayofweek_sin'] = np.sin(2 * np.pi * df_subset['date'].dt.dayofweek/7)
df_subset['dayofweek_cos'] = np.cos(2 * np.pi * df_subset['date'].dt.dayofweek/7)

df_subset.head()
```

```python
def add_cyclic_sin_cos_features(df, datecolumn = "date"):
    # Create sin and cos features for day of year
    df['dayofyear_sin'] = np.sin(2 * np.pi * df[datecolumn].dt.dayofyear/365.25)
    df['dayofyear_cos'] = np.cos(2 * np.pi * df[datecolumn].dt.dayofyear/365.25)

    # Create sin and cos features for month
    df['month_sin'] = np.sin(2 * np.pi * df[datecolumn].dt.month/12)
    df['month_cos'] = np.cos(2 * np.pi * df[datecolumn].dt.month/12)

    # Create sin and cos features for day of week
    df['dayofweek_sin'] = np.sin(2 * np.pi * df[datecolumn].dt.dayofweek/7)
    df['dayofweek_cos'] = np.cos(2 * np.pi * df[datecolumn].dt.dayofweek/7)

    return df
```
